```python
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()

# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Verify that Alpaca keys were loaded
print(f"Alpaca key type:{type(alpaca_api_key)}")
print(f"Alpaca secret tyep: {type(alpaca_secret_key)}")
```

```python
# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
### DATA COLLECTION ###
```

```python
# Subset your tickers, then pull returns data:
# Set the ticker
ticker = ["SPY", "AGG"]

# Set timeframe to '1D'
timeframe = "1D"

# Set start and end datetimes of 1 year, between now and 365 days ago.
# Results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2017-01-01', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2017-12-31', tz='America/New_York').isoformat()

# Get 1 year's worth of historical data for SPY and AGG
df = api.get_barset(
    ticker,
    timeframe,
    limit=None,
    start=start_date,
    end=end_date,
    after=None,
    until=None,
).df

# Drop unnecessary columns
df = df.drop(
    columns=['open', 'high', 'low', 'volume'],
    level=1
)
```

```python
# Display sample data
df.head()
```

```python
### MONTE CARLO SIMULATION ###
```

```python
# Calculate the daily ROI and display a sample
daily_returns = df.pct_change().dropna()

daily_returns.head()
```

```python
# Computing volitility
daily_volatility = daily_returns.std()
spy_volatility = daily_volatility["SPY"]["close"]
agg_volatility = daily_volatility["AGG"]["close"]
```

```python
# Saving the last day's closing price
spy_last_price = df["SPY"]["close"][-1]
agg_last_price = df["AGG"]["close"][-1]
```

```python
# Seting up the Monte Carlo Parameters
number_simulations = 500
number_records = 252 * 30
monte_carlo = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation
for x in range(number_simulations):
    
    print(f"Running Simulation {x}...")
    
    # Creating initial simulated prices array 
    spy_prices = [spy_last_price]
    agg_prices = [agg_last_price]
    
    # Simulating 20-year returns
    for _ in range(number_records):
        spy_prices.append(
            spy_prices[-1] * (1 + np.random.normal(daily_returns.mean()["SPY"]["close"], spy_volatility)))
        agg_prices.append(
            agg_prices[-1] * (1 + np.random.normal(daily_returns.mean()["AGG"]["close"], agg_volatility)))

    # Creating DF of simulated prices
    retirement_portfolio = pd.DataFrame({
        "SPY Simulated Prices": spy_prices,
        "AGG Simulated Prices": agg_prices
    })

    # Calculating Portfolio Daily Returns
    retirement_portfolio_returns=retirement_portfolio.pct_change()
    
    # Setting Portfolio Weights
    stocks_wt = 0.60
    bonds_wt = 0.40
    
    
    # Calculating weighted portfolio returns 
    retirement_portfolio_returns = stocks_wt * retirement_portfolio_returns["SPY Simulated Prices"] + bonds_wt * retirement_portfolio_returns["AGG Simulated Prices"]
    
    # Calculating normalized cumulative return series
    monte_carlo[x] = (1 + retirement_portfolio_returns.fillna(0)).cumprod()
```

```python
# Checking simulation ran successfully
monte_carlo.head()
```

```python
# Visualize the Simulation
monte_carlo.plot(
    legend=None,
    title="Retirement Portfolio Simulations"
)
```

```python
# Selecting last row for the cumulative returns
ending_returns = monte_carlo.iloc[-1, :]
ending_returns = monte_carlo.tail(1).T
ending_returns.head()
```

```python
# Displaying the 90% confidence interval for the ending returns
con_int = ending_returns.quantile(q=[0.05, 0.95], numeric_only=True)
con_int
```

```python
# Visualizing distribution of the ending returns
plt.figure();
ending_returns.plot.hist(bins=50)
plt.axvline(con_int.iloc[0, 0], color="r")
plt.axvline(con_int.iloc[1, 0], color="r")
```

```python
## WITH $20K Initial investment, expected portfolio returns in dollars 
## at percentiles stated above
```

```python
# Setting initial $20,000 investment
initial_investment = 20000
```

```python
# Computing expected portfolio return in dollar amounts
initial_investment * ending_returns.quantile([.01, .5, .9])
```

```python
## ASSUMING CURRENT PROJECTED INCOME FROM THE PLAID ANALYSIS, WILL A 4% WITHDRAW RATE
## FROM THE PORTFOLIO MEET OR EXCEED THAT VALUE AT THE 10th PERCENTILE?
```

```python
# Setting Plaid's projected income
plaid_proj_ann_inc =  7389
```

```python
# Calculating the 4% withdrawal for the ending return as retirement income
retirement_income = .04 * (initial_investment * ending_returns.quantile(.1)).iloc[0]
```

```python
# Determining if the retirement income meets or exceeds the current projected income with if/else statement
if retirement_income >= plaid_proj_ann_inc:
    print(f"You will have ${retirement_income:.2f} in retirement income which is sufficient")
else:
    print(f"You will have ${retirement_income:.2f} in retirement income which is not sufficient")
```
